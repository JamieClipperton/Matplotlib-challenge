```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
phar_data_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")
# Display the data table for preview
phar_data_df.head()
```

```python
# Checking the number of mice.
num_mice = phar_data_df["Mouse ID"].nunique()
num_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_mice = phar_data_df.loc[phar_data_df[["Mouse ID", "Timepoint"]].duplicated()==True]
dup_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
k403_data = phar_data_df.loc[phar_data_df["Mouse ID"]=="k403"]
k403_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
nondup_phar_df = phar_data_df.drop_duplicates(["Mouse ID","Timepoint"])
nondup_phar_df
```

```python
# Checking the number of mice in the clean DataFrame.
nondup_phar_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
nondup_phar_df["Drug Regimen"].unique()
```

```python
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
drug_mn=nondup_phar_df.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
drug_med=nondup_phar_df.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
drug_var=nondup_phar_df.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
drug_std=nondup_phar_df.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
drug_erm=nondup_phar_df.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']
drug_sum=pd.DataFrame({"Mean": drug_mn, "Median": drug_med, "Variance": drug_var, "Standard Deviation": drug_std, "Standard Error of the Means": drug_erm})
drug_sum
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
alt_drug_sum=nondup_phar_df.groupby(["Drug Regimen"]).agg(['mean','median','var','std','sem'])["Tumor Volume (mm3)"]
alt_drug_sum
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
capo_df=nondup_phar_df.loc[nondup_phar_df["Drug Regimen"]=="Capomulin",:]
rami_df=nondup_phar_df.loc[nondup_phar_df["Drug Regimen"]=="Ramicane",:]
infu_df=nondup_phar_df.loc[nondup_phar_df["Drug Regimen"]=="Infubinol",:]
ceft_df=nondup_phar_df.loc[nondup_phar_df["Drug Regimen"]=="Ceftamin",:]
# Start by getting the last (greatest) timepoint for each mouse
capo_last=capo_df.groupby("Mouse ID").max()["Timepoint"]
capo_last_df=pd.DataFrame(capo_last)
capo_last_mrg=pd.merge(capo_last_df,nondup_phar_df,on=("Mouse ID","Timepoint"),how="left")
capo_last_mrg.head()

rami_last=rami_df.groupby("Mouse ID").max()["Timepoint"]
rami_last_df=pd.DataFrame(rami_last)
rami_last_mrg=pd.merge(rami_last_df,nondup_phar_df,on=("Mouse ID","Timepoint"),how="left")
rami_last_mrg.head()

infu_last=infu_df.groupby("Mouse ID").max()["Timepoint"]
infu_last_df=pd.DataFrame(infu_last)
infu_last_mrg=pd.merge(infu_last_df,nondup_phar_df,on=("Mouse ID","Timepoint"),how="left")
infu_last_mrg.head()

ceft_last=ceft_df.groupby("Mouse ID").max()["Timepoint"]
ceft_last_df=pd.DataFrame(ceft_last)
ceft_last_mrg=pd.merge(ceft_last_df,nondup_phar_df,on=("Mouse ID","Timepoint"),how="left")
ceft_last_mrg.head()

```

```python
# Capomulin quartiles, IQR, and outliers
capo_tum=capo_last_mrg["Tumor Volume (mm3)"]
capo_qtls=capo_tum.quantile([.25,.5,.75])
capo_lq=capo_qtls[0.25]
capo_uq=capo_qtls[0.75]
capo_iqr=capo_uq-capo_lq

capo_lb=capo_lq - (1.5*capo_iqr)
capo_ub=capo_uq + (1.5*capo_iqr)
print(f"The lower quartile of Capomulin tumors is :{capo_lq}")
print(f"The upper quartile of Capomulin tumors: {capo_uq}")
print(f"The interquartile range of Capomulin tumors: {capo_iqr}")
print(f"The median of Capomulin tumors: {capo_qtls[0.5]} ")
print(f"Values below {capo_lb} could be outliers.")
print(f"Values above {capo_ub} could be outliers.")

print("----------------------------------------------------------------")
#Ramicane quartiles,IQR, and outliers 
rami_tum=rami_last_mrg["Tumor Volume (mm3)"]
rami_qtls=rami_tum.quantile([.25,.5,.75])
rami_lq=rami_qtls[0.25]
rami_uq=rami_qtls[0.75]
rami_iqr=rami_uq-rami_lq

rami_lb=rami_lq - (1.5*rami_iqr)
rami_ub=rami_uq + (1.5*rami_iqr)
print(f"The lower quartile of Ramicane tumors is :{rami_lq}")
print(f"The upper quartile of Ramicane tumors: {rami_uq}")
print(f"The interquartile range of Ramicane tumors: {rami_iqr}")
print(f"The median of Ramicane tumors: {rami_qtls[0.5]} ")
print(f"Values below {rami_lb} could be outliers.")
print(f"Values above {rami_ub} could be outliers.")

print("----------------------------------------------------------------")

#Infubinol quartiles,IQR, and outliers 
infu_tum=infu_last_mrg["Tumor Volume (mm3)"]
infu_qtls=infu_tum.quantile([.25,.5,.75])
infu_lq=infu_qtls[0.25]
infu_uq=infu_qtls[0.75]
infu_iqr=infu_uq-infu_lq

infu_lb=infu_lq - (1.5*infu_iqr)
infu_ub=infu_uq + (1.5*infu_iqr)
print(f"The lower quartile of Infubinol tumors is :{infu_lq}")
print(f"The upper quartile of Infubinol tumors: {infu_uq}")
print(f"The interquartile range of Infubinol tumors: {infu_iqr}")
print(f"The median of Infubinol tumors: {infu_qtls[0.5]} ")
print(f"Values below {infu_lb} could be outliers.")
print(f"Values above {infu_ub} could be outliers.")

print("----------------------------------------------------------------")

#Ceftamin quartiles,IQR, and outliers
ceft_tum=ceft_last_mrg["Tumor Volume (mm3)"]
ceft_qtls=ceft_tum.quantile([.25,.5,.75])
ceft_lq=ceft_qtls[0.25]
ceft_uq=ceft_qtls[0.75]
ceft_iqr=ceft_uq-ceft_lq

ceft_lb=ceft_lq - (1.5*infu_iqr)
ceft_ub=ceft_uq + (1.5*infu_iqr)
print(f"The lower quartile of Ceftamin tumors is :{ceft_lq}")
print(f"The upper quartile of Ceftamin tumors: {ceft_uq}")
print(f"The interquartile range of Ceftamin tumors: {ceft_iqr}")
print(f"The median of Ceftamin tumors: {ceft_qtls[0.5]} ")
print(f"Values below {ceft_lb} could be outliers.")
print(f"Values above {ceft_ub} could be outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
drug_plt_data= [ capo_tum, rami_tum, infu_tum, ceft_tum]
drugs= ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

fig1, ax1 = plt.subplots(figsize=(9,9))
ax1.set_title("Tumor Volume at Selected Mouse",fontsize =12)
ax1.set_ylabel("Final Tumor Volume (mm3)", fontsize = 10)
ax1.set_xlabel("Drug Regimen", fontsize =10)
ax1.boxplot(drug_plt_data, labels=drugs, patch_artist=True, vert=True)
plt.ylim(10,80)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
m914_df=capo_df.loc[capo_df["Mouse ID"] == "w914",:]
fig2, ax2=plt.subplots(figsize=(9,9))
m914_tp=m914_df["Timepoint"]
m914_tv=m914_df["Tumor Volume (mm3)"]
plt.title("Tumor Volume vs Timepoint for Mouse w914 with Capomulin")
plt.plot(m914_tp, m914_tv,linewidth=3,marker="s",color="purple")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capo_df_avg=capo_df.groupby(["Mouse ID"]).mean()
plt.scatter(capo_df_avg["Weight (g)"],capo_df_avg["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Average Tumor Volume vs Mouse Weight for Capomulin")
plt.show()
```

## Correlation and Regression

```python
#Calculate the correlation coefficient and linear regression model 
capo_correl=st.pearsonr(capo_df_avg["Weight (g)"],capo_df_avg["Tumor Volume (mm3)"])
print("Correlation coefficient between mouse weight and average tumor volume is "+str(round(capo_correl[0],3)))
```

```python
# for mouse weight and average tumor volume for the Capomulin regimen
xvalv=capo_df_avg["Weight (g)"]
yvalv=capo_df_avg["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(xvalv,yvalv)
regress_val=xvalv*slope+intercept
line_eq="Y="+str(round(slope,2))+"X+"+str(round(intercept,2))
fig3,ax3=plt.subplots(figsize=(9,7))
plt.title("Average Tumor Volume vs Mouse Weight for Capomulin")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.annotate(line_eq,(20,30),color="red")
plt.plot(xvalv,regress_val,"r-")
plt.scatter(capo_df_avg["Weight (g)"],capo_df_avg["Tumor Volume (mm3)"])
print(f"The r-squared is : {rvalue**2}")
plt.show()
```
